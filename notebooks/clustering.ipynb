```python
df.head()
```

```python
sample = df.sample(frac=0.4)
with ProgressBar():
    sample = sample.compute()
```

```python
sample.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=100)
predictions = kmeans.fit_predict(sample)
predictions
```

```python
sample['class'] = predictions
class_table = sample.groupby('class').sum()
class_table = class_table.replace(to_replace=0, value=np.NAN).dropna(how='all', axis='columns').fillna(0)
class_table.head()
```

```python
normalized = class_table.div(class_table.sum(axis=1), axis=0)
columnar = pd.DataFrame(normalized.stack()).reset_index().rename(columns={0: 'proportion', 'level_1': 'symbol'})
columnar.head()
```
